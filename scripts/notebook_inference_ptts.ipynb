```python
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = ParlerTTSForConditionalGeneration.from_pretrained("/home/eblict/parler-tts-finetune/parler-tts/output_dir_vertex_male1_finetune", torch_dtype=torch.float16).to(device)
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/indic-parler-tts")

prompt = "সৌদি থেকে এল ১ লাখ টন তেল, পুরোদমে উৎপাদনে ফিরছে ইস্টার্ন রিফাইনারি"
description = "Sazzad speaks quite slowly with a moderate tone, offering a somewhat clear audio quality in a somewhat confined space with minimal background noise."

input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
prompt_input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)

generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
audio_arr = generation.cpu().numpy().squeeze()
```

```python
from IPython.display import Audio
Audio(audio_arr, rate=model.config.sampling_rate)
```
